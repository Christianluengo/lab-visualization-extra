## Coding Session I

From the Fitbit data, we want to visually understand: 

1. **How the average number of steps change by month. Use the appropriate visualization to show the median steps by month?**
1. **Is Fitbitter more active on weekend or workdays?**

##### All plots must be in the same jupyter notebook cell.

Hints:

- Explore `plt.xticks`
- Save your figures

```python
fitbit['Date'] = pd.to_datetime(fitbit['Date'], format=  "%Y-%m-%d")

mean_month = fitbit.groupby([(fitbit.Date.dt.month)])['Steps'].mean().reset_index()
meses = [datetime.date(2000, m, 1).strftime('%B') for m in range(1, 13)]
fig2 = sns.lineplot(data=mean_month, x="Date", y="Steps",color = "green", marker = "o", markersize = 7, linewidth = 1)
fig2.set_xticklabels(meses, rotation = 45)
                     
plt.show();
```

## Coding Session II 

**Write a loop to plot 3 scatter plots of the following features:**

1. **`Minutes Lightly Active`  vs `Steps`**
1. **`Minutes Very Active` vs `Steps`**
1. **`Minutes Sedentary` vs `Steps`**

```python
def auto_grafico(x,y):
    sns.scatterplot(data = fitbit, x = f"{x}", y = f"{y}", color = "green")
    plt.title(f"{x} vs {y}")
    plt.show();  

# aunque la variable "y" sea fija no la definimos como tal para no limitar la funciÃ³n y poder analizar otras variables si fuera necesario
```

```python
auto_grafico('Minutes Lightly Active', 'Steps')
auto_grafico('Minutes Very Active', 'Steps')
auto_grafico('Minutes Sedentary', 'Steps')
```
